Experiment 1

Aim: Implement Multilayer Perceptron algorithm to simulate XOR gate.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# XOR input dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# XOR output dataset
y = np.array([[0], [1], [1], [0]])
```

```python
# Define the model
model = Sequential()

# Input layer with 2 inputs, and hidden layer with 2 neurons using the 'tanh' activation function
model.add(Dense(2, input_dim=2, activation='tanh'))

# Output layer with 1 neuron using the 'sigmoid' activation function
model.add(Dense(1, activation='sigmoid'))

# Compile the model using binary cross-entropy as the loss function
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X, y, epochs=5000, verbose=0)
```

```python
# Test the model on the XOR inputs
predictions = model.predict(X)

# Print the results
print("Predicted outputs:")
print(np.round(predictions))
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X, y)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```
